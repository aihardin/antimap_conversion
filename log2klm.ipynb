```python
import glob
#import pandas as pd
import gzip
import os
import zipfile
try:
    import zlib
    compression = zipfile.ZIP_DEFLATED
except:
    compression = zipfile.ZIP_STORED

modes = { zipfile.ZIP_DEFLATED: 'deflated',
          zipfile.ZIP_STORED:   'stored',
          }

import csv
import urllib2
from datetime import datetime,timedelta
from lxml import etree
from pykml.factory import KML_ElementMaker as KML

# csv format
# file name is: MMDDYY_HHMM_SS.csv.gz
# line data is: lat, long, compass, speed(kph), distance (kms), time (milliseconds), 
# and finally an input field for optional points of interest.
# other reference: 
# http://www.doughellmann.com/PyMOTW/zipfile/
# https://developers.google.com/kml/documentation/kmzarchives

def millisec_to_time(ms):
    d = 0
    s = 0
    while ms >= 1000:
        s += 1
        ms -= 1000
        if s >= 3600*24:
            d += 1
            s -= 3600*24
    return timedelta(days=d,seconds=s,milliseconds=ms)    
    
```

```python
#os.chdir('2014')
log_files = glob.glob('*.csv.gz')
anti_map_headers = ['latitude','longitude', 'bearing', 'speed', 'distance', 'time','label']
DOWNSAMPLE = 300 # logged at 30 points/sec so just keep one per 10 second
for log_file in log_files:
    
    doc = KML.Document()
     
    doc.append(KML.Style(
                KML.ListStyle(
                  KML.listItemType('checkHideChildren'),
                  ),
                id='check-hide-children',
                )
    )
    doc.append(KML.styleUrl('check-hide-children'),)
    doc.append(KML.Style(
                KML.LineStyle(
                  KML.color('8000ffff'),
                  KML.width(6),
                  ),
                 KML.PolyStyle(
                               KML.color('60ff0000')),
                id='line',
                )
    )
    style_id = 'shaded_dot'
    doc.append(KML.Style(
                KML.IconStyle(
                    KML.Icon(KML.href("http://maps.google.com/mapfiles/kml/paddle/ylw-blank-lv.png")),
                    #KML.Icon(KML.href("http://maps.google.com/mapfiles/kml/shapes/shaded_dot.png")),
                    #KML.Scale('x="1" y="1" xunits="pixels" yunits="pixels"')
                    KML.Scale('0.5')
                ),
                id=style_id,
                )
    )
    
    #doc.append(KML.Folder())
    print log_file
    time_start = datetime.strptime(log_file.split('.')[0], '%d%m%y_%H%M_%S')
    out_stamp = time_start.strftime('%y%m%d_%H%M_%S')
    kml_file = out_stamp+'.kml'
    #print kml_file
    #break
    
    print 'log start:', time_start
    
    #kml_folder = kml_file.split('.')[0]
    kml_folder = './'
    if os.path.exists(kml_file):
        print kml_file, 'exists'
        continue
    
    print kml_file
    log_fh = gzip.open(log_file,'rb')
    idx = -1
    coordinates = ''
    for row in csv.DictReader(log_fh,fieldnames=anti_map_headers):
        idx += 1
        if idx % DOWNSAMPLE != 0:
            continue
        row_time = millisec_to_time(int(row['time']))
        timestamp = time_start + row_time
        coordinates += '''%s,%s ''' % (row["longitude"],row["latitude"])
        #coordinates += '\n'
        #print 'timepoint', timestamp
        #pm = KML.Placemark(
        #                   KML.TimeStamp(KML.when(timestamp.strftime('%Y-%m-%dT%H:%M:%SZ')),),
        #                   KML.Point(KML.coordinates("{0},{1}".format(row["longitude"],row["latitude"]))),
        #                   KML.styleUrl('#{0}'.format(style_id)),
        #                 )
        #doc.Folder.append(pm)
        #doc.append(pm)
    pm = KML.Placemark(
                       KML.LineString(
                                      KML.coordinates(coordinates)
                                      ),
                        KML.styleUrl('#line')
                    )
    doc.append(pm)
    log_fh.close()
    
    
    kml_path = os.path.join(kml_folder,kml_file)    
    kml_writer = open(kml_path,'w')
    kml_writer.write(etree.tostring(doc, pretty_print=True))
    kml_writer.close()
    
    #print 'creating archive'
    kmz_file = out_stamp+'.kmz'
    zf = zipfile.ZipFile(kmz_file, mode='w')
    try:
        print 'adding %s with compression mode' % (kml_path), modes[compression]
        zf.write(kml_path, compress_type=compression)
    finally:
        #print 'closing'
        zf.close()
    
    
    
    
```
